```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# ...existing code...
from pathlib import Path

# desired canonical columns
desired = ['Open', 'High', 'Low', 'Close', 'Volume']
canonical = {c.lower(): c for c in desired}

# map existing columns case-insensitively
cols_lower = {col.lower(): col for col in df.columns}
selected_cols = []
missing = []
for lname, canon in canonical.items():
    if lname in cols_lower:
        selected_cols.append(cols_lower[lname])
    else:
        missing.append(canon)

if missing:
    print("Warning - these desired columns were not found and will be missing:", missing)

if not selected_cols:
    raise RuntimeError("No desired columns found in df. Check column names.")

# select and standardize names
df = df[selected_cols].copy()
df.columns = [canonical[col.lower()] for col in df.columns]

# ensure index is datetime (if Date is a column instead, set it first)
if not isinstance(df.index, pd.DatetimeIndex):
    try:
        df.index = pd.to_datetime(df.index)
    except Exception as e:
        print("Index could not be converted to datetime:", e)

# save cleaned ohlcv csv
out_path = Path.cwd() / "data" / "AAPL_ohlcv.csv"
out_path.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(out_path, index=True)
print(f"Saved cleaned OHLCV to: {out_path}")
# ...existing code...
```
